```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#4 layers Training Data vs Training Accuracy
train_accuracy = [95.92, 98.28, 96.69, 97.44]
train_data = [20000, 40000, 60000, 73000]
```

```python
plt.figure(figsize=(8, 5))

plt.title('Training Data vs. Training Accuracy (4 Convolution Layers)') 
plt.xlabel('Training Data') 
plt.ylabel('Training Accuracy') 
plt.plot(train_data, train_accuracy, color ="red") 
plt.show()
```
