```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#Number of Layers vs Training Accuracy with 20k training data and multiple layers
train_accuracy = [95.53, 97.44, 98.99]
no_layers = [2, 4, 6]
```

```python
plt.figure(figsize=(8, 5))

plt.title('No of Convlution Layers vs. Training Accuracy (20000 Training Data') 
plt.xlabel('Number of Layers') 
plt.ylabel('Training Accuracy') 
plt.plot(no_layers, train_accuracy, color ="red") 
plt.show()
```

```python
#Number of Layers vs Validation Accuracy with 20k training data and multiple layers
val_accuracy = [93.77, 93.42, 94.52]
no_layers = [2, 4, 6]
```

```python
plt.figure(figsize=(8, 5))

plt.title('No of Convlution Layers vs. Validation Accuracy (20000 Training Data') 
plt.xlabel('Number of Layers') 
plt.ylabel('Validation Accuracy') 
plt.plot(no_layers, val_accuracy, color ="red") 
plt.show()
```
