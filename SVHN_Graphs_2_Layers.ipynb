```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#2 layers Training Data vs Training Accuracy
train_accuracy = [94.39, 94.97, 95.27, 95.53]
train_data = [20000, 40000, 60000, 73000]
```

```python
plt.figure(figsize=(8, 5))

plt.title('Training Data vs. Training Accuracy (2 Convolution Layers)') 
plt.xlabel('Training Data') 
plt.ylabel('Training Accuracy') 
plt.plot(train_data, train_accuracy, color ="red") 
plt.show()
```

```python
#2 layers Training Data vs Validation Accuracy
val_accuracy = [91.22, 92.99, 93.74, 93.77]
train_data = [20000, 40000, 60000, 73000]
```

```python
plt.figure(figsize=(8, 5))

plt.title('Training Data vs. Validation Accuracy (2 Convolution Layers)') 
plt.xlabel('Training Data') 
plt.ylabel('Validation Accuracy') 
plt.plot(train_data, val_accuracy, color ="red") 
plt.show()
```
