```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#6 layers Training Data vs Training Accuracy
train_accuracy = [98.38, 98.90, 99.01, 98.99]
train_data = [20000, 40000, 60000, 73000]
```

```python
plt.figure(figsize=(8, 5))

plt.title('Training Data vs. Training Accuracy (6 Convolution Layers)') 
plt.xlabel('Training Data') 
plt.ylabel('Training Accuracy') 
plt.plot(train_data, train_accuracy, color ="red") 
plt.show()
```

```python
#6 layers Training Data vs Validation Accuracy
val_accuracy = [92.60, 93.36, 94.55, 94.52]
train_data = [20000, 40000, 60000, 73000]
```

```python
plt.figure(figsize=(8, 5))

plt.title('Training Data vs. Validation Accuracy (6 Convolution Layers)') 
plt.xlabel('Training Data') 
plt.ylabel('Validation Accuracy') 
plt.plot(train_data, val_accuracy, color ="red") 
plt.show()
```
